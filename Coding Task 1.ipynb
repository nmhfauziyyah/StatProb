```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
file_excel = 'Dataset (1).xlsx'
data = pd.read_excel(file_excel)
```

```python
PA = data ['Penjualan A (pcs)']
```

```python
mean = np.mean(PA)
median = np.median(PA)
mode = stats.mode(PA)
range_value = np.ptp(PA)  # range
iqr = stats.iqr(PA)  # Interquartile Range (IQR)
variance = np.var(PA, ddof=1)  # Variance
std_dev = np.std(PA, ddof=1)  # Standard deviation
skewness = stats.skew(PA)  # Skewness
kurtosis = stats.kurtosis(PA)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Range: {range_value}")
print(f"IQR: {iqr}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurtosis}")
```

```python
import pandas as pd
import numpy as np

PA = data ['Penjualan A (pcs)']

# Membuat tabel distribusi frekuensi
bins = np.histogram_bin_edges(PA, bins='auto')  # Membuat interval otomatis
frequency, bin_edges = np.histogram(PA, bins=bins)  # Menghitung frekuensi untuk tiap interval

# Menyusun tabel distribusi frekuensi
freq_table = pd.DataFrame({
    'Interval': pd.IntervalIndex.from_breaks(bin_edges),
    'Frekuensi': frequency
})

# Menampilkan tabel distribusi frekuensi
print(freq_table)
```

```python
import matplotlib.pyplot as plt

# Membuat histogram
plt.hist(PA, bins=bins, edgecolor='black')

# Menambahkan judul dan label
plt.title('Histogram Penjualan A')
plt.xlabel('Penjualan A (pcs)')
plt.ylabel('Frekuensi')

# Menampilkan histogram
plt.show()
```

```python
# Menghitung titik tengah dari setiap bin
bin_centers = 0.5 * (bin_edges[1:] + bin_edges[:-1])

# Membuat poligon frekuensi
plt.plot(bin_centers, frequency, marker='o', linestyle='-', color='blue')

# Menambahkan judul dan label
plt.title('Poligon Frekuensi Penjualan A')
plt.xlabel('Penjualan A (pcs)')
plt.ylabel('Frekuensi')

# Menampilkan poligon frekuensi
plt.show()
```
